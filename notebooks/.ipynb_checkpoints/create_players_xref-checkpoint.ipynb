```python
# Setup the starting dataframe

con_memory = duckdb.connect(database=':memory:')
players = construct_df_roster(con_memory)
df = players[(players['position'] == 'RB')].sort_values(by=['name', 'season']).reset_index(drop=True)

print(len(df.name.unique().tolist()), 'players')
print(df[(df.pfr_id.isnull())].shape[0], 'nulls')

df['pfr_id'] = df.groupby('name')['pfr_id'].transform(
    lambda x: x.ffill().bfill())

df = df[['name', 'pfr_id']].dropna().drop_duplicates('pfr_id').reset_index(drop=True)

print(len(df.name.unique().tolist()), 'players')
print(df[(df.pfr_id.isnull())].shape[0], 'nulls')

df = players[(players.name.isin(df.name.unique().tolist())) & (players.pfr_id.isin(df.pfr_id.unique().tolist()))]
print(len(df.name.unique().tolist()), 'players')
print(df[(df.pfr_id.isnull())].shape[0], 'nulls')

# print(len(df.name.unique().tolist()), 'players')
# con_memory.register('df_og', df)
# df = con_memory.execute("SELECT * FROM df WHERE name IN (SELECT name FROM df WHERE pfr_id IS NULL) AND pfr_id IS NOT NULL").fetchdf()
# print(len(df.name.unique().tolist()), 'players')
# df.to_csv("../tables/test.csv")

# custom_add = pd.read_csv("../src/rosters/roster_2009.csv")
# custom_add = custom_add[(custom_add['full_name'] == "Ricky Williams")].fillna({'birth_date': '01/01/1990', 'pfr_id': 'UNKNOWN'})
# custom_add['depth_chart_position'] = 'RB'
# df = pd.concat([df, custom_add])
display(df)
con_memory.close()
```
